```python
def extended_gcd(a, b):  
    if b == 0:  
        return a, 1, 0  
    else:  
        gcd, x1, y1 = extended_gcd(b, a % b)  
        x = y1  
        y = x1 - (a // b) * y1  
        return gcd, x, y   
```

```python
def reverse_value(a, m, func):
    NOD, u, _ = func(a, m)
    if(NOD == 1):
        return u%m
    else:
        raise Exception
```

```python
def finding_e_d(e, p, q):
    d = reverse_value(e, (p-1)*(q-1), extended_gcd)
    if(extended_gcd(e, (p-1)*(q-1))[0] == 1):
        return (e, d)
    else:
        raise Exception
```

```python
def get_blocks(digits, n):
    """
    Разбивает строку цифр на блоки < n, без ведущих нулей.
    Пример: "121027419..." → [12102, 7419, ...]
    """
    blocks = []
    i = 0
    while i < len(digits):
        # Макс. возможная длина блока — сколько цифр в (n-1)
        max_len = len(str(n - 1))
        
        # Пробуем от max_len до 1
        block_len = min(max_len, len(digits) - i)
        found = False
        
        while block_len >= 1:
            candidate = digits[i:i + block_len]
            
            # Проверка 1: не начинается с '0' (кроме случая "0" — но у тебя коды ≥10, так что не будет)
            if len(candidate) > 1 and candidate[0] == '0':
                block_len -= 1
                continue
            
            num = int(candidate)
            
            # Проверка 2: num < n
            if num < n:
                blocks.append(num)
                i += block_len
                found = True
                break
            
            block_len -= 1
        
        if not found:
            raise ValueError(f"Не удалось разбить цифры '{digits[i:]}' на блоки < {n}")
    
    return blocks
```

```python
def pow_mod(base, exp, mod):
    result = 1
    base = base % mod
    while exp > 0:
        if exp & 1:
            result = (result * base) % mod
        base = (base * base) % mod
        exp >>= 1
    return result
```

```python
def cipher_decipher(messages, e, n):
    res = []
    for message in messages:
        res.append(pow_mod(message, e, n))
    return res
```
