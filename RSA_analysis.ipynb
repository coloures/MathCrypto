```python
def extended_gcd(a, b):  
    if b == 0:  
        return a, 1, 0  
    else:  
        gcd, x1, y1 = extended_gcd(b, a % b)  
        x = y1  
        y = x1 - (a // b) * y1  
        return gcd, x, y   
```

```python
def reverse_value(a, m, func):
    NOD, u, _ = func(a, m)
    if(NOD == 1):
        return u%m
    else:
        raise Exception("Обратного элемента не существует")
```

```python
def finding_e_d(e, p, q):
    d = reverse_value(e, (p-1)*(q-1), extended_gcd)
    if(extended_gcd(e, (p-1)*(q-1))[0] == 1):
        return (e, d)
    else:
        raise Exception
```

```python
def pow_mod(base, exp, mod):
    result = 1
    base = base % mod
    while exp > 0:
        if exp & 1:
            result = (result * base) % mod
        base = (base * base) % mod
        exp >>= 1
    return result
```

```python
def decipher(messages, e, n):
    res = []
    for message in messages:
        res.append(pow_mod(message, e, n))
    return res
```

```python
def factorization(n):
    for p in range(2, int(math.isqrt(n)) + 1):
        if n % p == 0:
            q = n // p
            return p, q
    raise ValueError(f"Число {n} простое или не удалось факторизовать")
```
