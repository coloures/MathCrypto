```python
import math
```

```python
def build_sieve(p, m):
    sieve = [False] * p
    for x in range(p):
        z = (x*x - m) % p
        if z == 0:
            sieve[x] = True
        else:
            if(pow(z, (p-1)//2, p) == 1): # –ø—Ä–æ–≤–µ—Ä–∫–∞ —á–µ—Ä–µ–∑ –∫—Ä–∏—Ç–µ—Ä–∏–π –≠–π–ª–µ—Ä–∞ 
                sieve[x] = True
    return sieve
```

```python
def overlaying_sieves(m, sieves):
    final_sieve = []
    for x in range(math.isqrt(m)+1, (m+1) // 2 + 1):
        final_sieve.append(x)
    for sieve in sieves:
        start = (math.isqrt(m)+1)%len(sieve)
        for i in range(len(final_sieve)):
            final_sieve[i] = -1 if sieve[(start+i)%len(sieve)] == False else final_sieve[i]
    return final_sieve
```

```python
def is_perfect_square_simple(n):
    if n < 0:
        return False, None
    y = math.isqrt(n)
    if y * y == n:
        return y
    else:
        return None
```

```python
def get_p_and_q(x, y):
    return x+y, x-y
```

```python
def extended_gcd(a, b):  
    if b == 0:  
        return a, 1, 0  
    else:  
        gcd, x1, y1 = extended_gcd(b, a % b)  
        x = y1  
        y = x1 - (a // b) * y1  
        return gcd, x, y   
```

```python
def p_method(m, x_1, x_2):
    step = 0
    while True:
        step += 1
        x_1 = (x_1*x_1 + 1) % m
        x_2 = (x_2*x_2 + 1) % m
        x_2 = (x_2*x_2 + 1) % m
        #print(f"{x_1} {x_2} —à–∞–≥ {step}")
        d = extended_gcd(abs(x_1 - x_2), m)[0]
        if d > 1:
            break
        elif d == m:
            return Exception("–ù–µ —É–¥–∞–µ—Ç—Å—è –Ω–∞–π—Ç–∏ –ø—Ä–æ—Å—Ç–æ–π –¥–µ–ª–∏—Ç–µ–ª—å –ø–æ–º–∏–º–æ 1")
    return d, int(m/d), step
```
