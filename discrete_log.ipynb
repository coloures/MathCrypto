```python
def isqrt(p):
    x = p
    while True:
        y = (x + int(p/x)) >> 1
        if(y < x):
            x = y
        else:
            return x
```

```python
def pow_mod(base, exp, mod):
    result = 1
    base = base % mod
    while exp > 0:
        if exp & 1:
            result = (result * base) % mod
        base = (base * base) % mod
        exp >>= 1
    return result
```

```python
get_k = lambda p: isqrt(p) + 1
```

```python
y = lambda a, n, k, p: pow_mod(a, n*k, p)
z = lambda b, a, n,p: (b *pow_mod(a, n, p))%p
```

```python
def print_table(y_s, z_s, k):
    print("n\t", end="")
    for i in range(1, min(k, len(y_s)) + 1):
        print(f"{i:>8}", end="")
    print()
    
    print("y_n\t", end="")
    for i in range(min(k, len(y_s))):
        print(f"{y_s[i]:>8}", end="")
    print()
    
    print("z_n\t", end="")
    for i in range(min(k, len(z_s))):
        print(f"{z_s[i]:>8}", end="")
    print()
```

```python
def get_x(a, b, p, max_iters):
    k = get_k(p)
    n = 1
    y_s = []
    z_s = []
    while True:
        y_current = y(a,n,k,p)
        z_current = z(b,a,n,p)
        y_s.append(y_current)
        z_s.append(z_current)
        for idx, yi in enumerate(y_s):
            if(yi in z_s):
                j = z_s.index(yi)
                x = (idx + 1) * k - (j + 1)
                if pow_mod(a, x, p) == b % p:
                    print_table(y_s, z_s, k)
                    return x, y_s, z_s, idx+1, j+1
        n += 1
        if(n > max_iters):
            return None, y_s, z_s
```
