```python
class SravneniyeError(Exception):
    pass
```

```python
class ReverseElementError(Exception):
    pass
```

Вычисление НОДа двух чисел

```python
def extended_gcd(a, b):  
    if b == 0:  
        return a, 1, 0  
    else:  
        gcd, x1, y1 = extended_gcd(b, a % b)  
        x = y1  
        y = x1 - (a // b) * y1  
        return gcd, x, y   
```

```python
extended_gcd(7, 4)
```

Поиск обратного числа

```python
def reverse_value(a, m, func):
    NOD, u, _ = func(a, m)
    if(NOD == 1):
        return u
    else:
        raise ReverseElementError()
```

Решение сравнения

```python
def sravneniye(a, b, m, NOD_func, Obratnoye_func):
    if(b % NOD_func(a,m)[0] != 0):
        raise SravneniyeError()
    else:
        NOD = NOD_func(a,m)[0]
        a //= NOD
        b //= NOD
        m //= NOD

        x = (b * Obratnoye_func(a, m, NOD_func))%m

        lst = [m * i + x for i in range(NOD)]

        return lst
```

```python
sravneniye(9, 6, 12, extended_gcd, reverse_value)
```

Решение системы сравнений

```python
def systema_sravneniy(a1, a2, b1, b2, m, NOD_func, Obratnoye_func):
    return [(x, (b1 - a1*x)%m) for x in sravneniye(a1-a2, b1-b2, m, NOD_func, Obratnoye_func)]
```

```python
systema_sravneniy(3, 7, 5, 1, 8, extended_gcd, reverse_value)
```

Кодирование аффинным шифром

```python
symbols = ['а','б','в','г','д','е','ж','з','и','й','к','л','м','н','о','п','р','с','т','у','ф','х','ц','ч','ш','щ','ъ','ы','ь','э','ю','я']
```

```python
def crypt(a, b, alfavit ,str_):
    return "".join([alfavit[(a*alfavit.index(x)+b)%len(alfavit)] for x in str_])
```

Декодирование аффинным шифром

```python
def decrypt(a, b, alfavit ,str_):
    return "".join([alfavit[(alfavit.index(y)-b)*a%len(alfavit)] for y in str_])
```
