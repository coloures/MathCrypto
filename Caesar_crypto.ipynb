```python
symbols = ['а','б','в','г','д','е','ж','з','и','й','к','л','м','н','о','п','р','с','т','у','ф','х','ц','ч','ш','щ','ъ','ы','ь','э','ю','я']
```

```python
print(len(symbols))
```

```python
def encrypt(str_, symbols, symbols_without_changing, forbidden_symbols, k, c):
    encrypted = ""
    for i in range(len(str_)):
        if(str_[i] in symbols_without_changing):
            encrypted += str_[i]
        elif(str_[i] in forbidden_symbols):
            raise Exception("Нельзя такое сюда вводить")
        else:
            encrypted += symbols[(symbols.index(str_[i]) + k)%c]
    return encrypted
```

```python
def decrypt(str_, symbols, symbols_without_changing, c):
    for i in range(c):
        decrypted = ""
        for j in range(len(str_)):
            if(str_[j] in symbols_without_changing):
                decrypted += str_[j]
            else:
                decrypted += symbols[(symbols.index(str_[j]) - i)%c]
        while(True):
            res = input(f"При сдвиге равном {i} получается сообщение {decrypted}. Если оно правильно расшифровано - введите Right, иначе - Next. Затем нажмите Enter. ")
            if(res == "Right"):
                return decrypted
            elif(res == "Next"):
                break
            else:
                continue
```

```python
print(encrypt(input("Введите своё сообщение: "), symbols, [":", ".", ",", "!", " "], [], 7, len(symbols)))
```

```python
print(decrypt(input("Введите свой шифр: "), symbols, [":", ".", ",", "!", " "], len(symbols)))
```
