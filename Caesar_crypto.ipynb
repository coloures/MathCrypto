```python
symbols = ['а','б','в','г','д','е','ж','з','и','й','к','л','м','н','о','п','р','с','т','у','ф','х','ц','ч','ш','щ','ъ','ы','ь','э','ю','я']
```

```python
def encrypt(str_, symbols, symbols_without_changing, forbidden_symbols, k, c, changed_symbols =None):
    encrypted = ""
    for i in range(len(str_)):
        if(changed_symbols != None and str_[i].lower() in changed_symbols.keys()):
            current = changed_symbols[str_[i].lower()]
        else:
            current = str_[i].lower()
        if(current in symbols_without_changing):
            encrypted += current
        elif(current in forbidden_symbols):
            raise Exception("Нельзя такое сюда вводить")
        else:
            encrypted += symbols[(symbols.index(current) + k)%c]
    return encrypted
```

```python
def decrypt(str_, symbols, symbols_without_changing, c, file_ = None, changed_symbols = None):
    for i in range(c):
        decrypted = ""
        for j in range(len(str_)):
            if(changed_symbols != None and str_[j].lower() in changed_symbols.keys()):
                current = changed_symbols[str_[j].lower()]
            else:
                current = str_[j].lower()
            if(current in symbols_without_changing):
                decrypted += current
            else:
                decrypted += symbols[(symbols.index(current) - i)%c]
        while(True):
            res = input(f"При сдвиге равном {i} получается сообщение {decrypted}. Если оно правильно расшифровано - введите 1, иначе - 2. Затем нажмите Enter. ")
            if(file_ != None):
                try:
                    file_.write(f"{decrypted} k = {i}" + '\n')
                except Exception:
                    file_.close()
            if(res == "1"):
                return decrypted
            elif(res == "2"):
                break
            else:
                continue
```
